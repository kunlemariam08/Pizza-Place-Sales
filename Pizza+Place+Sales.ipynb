```python
#Q6. What are the peak hours of sales?


# Convert 'time' column to datetime format
df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S')

# Extract the hour from each order
df['hour'] = df['time'].dt.hour

# Count number of orders per hour
peak_hours = df['hour'].value_counts().sort_index()

# Display the result
print(peak_hours)
```

```python
# Q7. Find the total sales made on each day of the week. Which day of the week is when sales are made the mos
# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract day of the week
df['day_of_week'] = df['date'].dt.day_name()

# Calculate total sales per row
df['sales'] = df['price'] * df['quantity']

# Group by day of the week and sum sales
sales_by_day = df.groupby('day_of_week')['sales'].sum().sort_values(ascending=False)

# Display results
print(sales_by_day)
```

```python
top_day = sales_by_day.idxmax()
top_sales = sales_by_day.max()
print(f"The highest sales were on {top_day} with total sales of ${top_sales:.2f}")
```

```python
# Q8. Find the top 5 bestselling pizzas
top_pizzas = df.groupby('name')['quantity'].sum().sort_values(ascending=False).head(5)
print("Top 5 bestselling pizzas:")
print(top_pizzas)
```

```python
#Q9. Find the sales made in each month. Any trend noticeable?
# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract month name
df['month'] = df['date'].dt.month_name()

# Calculate total sales per row
df['sales'] = df['price'] * df['quantity']

# Group by month and sum sales
monthly_sales = df.groupby('month')['sales'].sum().sort_values(ascending=False)

# Display results
print(monthly_sales)
```

```python
#Q10. Are there pizza types that are not doing well on the menu?
# Group by pizza type and sum quantity sold
pizza_type_sales = df.groupby('pizza_type_id')['quantity'].sum().sort_values()

# Display the bottom performers
print("Least performing pizza types:")
print(pizza_type_sales.head(5))
```
